```python
import requests
from bs4 import BeautifulSoup

infos = []
intro = []
res = []

for page in range(1, 100):
    url = 'https://movies.yahoo.com.tw/movieinfo_main/' + str(page)
    print(url)
    
    r = requests.get(url)
    soup = BeautifulSoup(r.text)
    
    if soup.find('title').text == 'Yahoo奇摩電影':
        continue
    
    info = {}
    for d in soup.find_all('div', class_="movie_intro_info_r"):

        info['名稱'] = d.find('h1').text.replace(' ', '').replace('\n', '')

        category = d.find('div', class_="level_name_box")
        category_item = category.find_all('a',class_="gabtn")
        tmp_item = str("")
        for item in category_item:
            tmp_item += item.text.replace(' ', '').replace('\n', '') + '.'
        info['分類'] = tmp_item
        
        info['上映日期'] = d.find('span').text.replace('上映日期：', '')
        
    intr = {}
    intr['劇情介紹'] = soup.find('span', id='story').text.replace(' ', '').replace('\n', '')
    
    infos.append(info)
    intro.append(intr)
    
for i, j in list(zip(infos, intro)):
    print(i)
    res.append(dict(list(i.items()) + list(j.items())))

res
```

```python
#利用 DF 存取 / 另存成 CSV

import pandas as pd

df = pd.DataFrame(res)
df.to_csv('HW#2.csv', encoding='utf_8_sig')

df
```
