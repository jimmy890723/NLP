```python
def sum(*a):
    print('=> ', a)
    t = 0
    for i in a:
        t = t + i 
    return t

print(sum(1, 2)) # 3
print(sum(1, 2, 3)) # 6
```

```python
l = [1, 2, 3, 4, 5, 6, 7, 8, 9, 0]


evens = []
odds = []

for i in l:
    # print(i, i % 2)
    if i % 2 == 0:
        # print(i, ' => 偶數')
        evens.append(i)
    if i % 2 == 1:
        # print(i, ' => 奇數')    
        odds.append(i)
        
l = evens + odds
print(l)
```

```python
S = "Please count the character here."
d = {} # dict()

# 1. 先把每個字母掃過一次
for s in S:
    # print(s)
    # 2. 檢查字母有沒有存在表格中
    if s in d: # 有存在 => 累加
        d[s] = d[s] + 1
    if s not in d: # 沒有存在 => 新增
        d[s] = 1

print(d)
```

```python
# 參考解答：④ 請問 Dcard 總共有幾個看板？
import requests
r = requests.get('https://www.dcard.tw/_api/forums')
response = r.text

import json
data = json.loads(response)

len(data)
```

```python
# 參考解答：⑤ 請問最熱門的看板是哪一個？ - 方法一

MAX = 0
NAME = ''

for d in data:
    # print(d['name'], d['subscriptionCount'])
    if d['subscriptionCount'] > MAX:
        MAX = d['subscriptionCount']
        NAME = d['name']
          
print(NAME, MAX)
```

```python
# 參考解答：⑤ 請問最新建立的看板是哪一個？

# sortData = sorted(data, key=lambda k: k['createdAt'])

def f(d):
    return d['createdAt']

sortData = sorted(data, key=f)

for d in sortData[::-1]:
    print(d['name'], d['createdAt'])
```
